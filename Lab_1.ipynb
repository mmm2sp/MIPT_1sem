```python
import numpy as np
import matplotlib.pyplot as plt
U_V = [0, 0.1, 0.13, 0.155, 0.195, 0.45, 0.325, 0.235, 0.195, 0.16, 0.125]
I2_mA = [0, 48.2 ** 2, 61.1 ** 2, 72.6 ** 2, 94.3 ** 2, 212.4 ** 2, 157.0 ** 2, 111.2 ** 2, 93.6 ** 2, 76.3 ** 2, 60.1 ** 2]
a_1 ,e_1 = np.polyfit(U_V, I2_mA, deg=1, cov=True)
a_2 ,e_2 = np.polyfit(U_V, I2_mA, deg=2, cov=True)
x = np.arange(0, 0.45, 0.01)
p_1 = np.poly1d(a_1)
p_2 = np.poly1d(a_2)
plt.plot(x, p_1(x))
plt.plot(x, p_2(x))
plt.errorbar(U_V, I2_mA, xerr=0.007, yerr=0.1, marker='^', lw = 0)
plt.title(r'$I^2 (U)$ проволоки длиной $L = 20.0 $ см')
plt.ylabel(r'$I^2, mA^2$')
plt.xlabel(r'$U, V$')
plt.grid(True)
plt.show()
```
